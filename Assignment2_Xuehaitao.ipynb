```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('D:/00NTU_2020FALL/Bio_Statistics_Mathematics/Bio-statistic.csv')
```

```python
data_controls = data['Controls']
data_Pre_diabetics = data['Pre-diabetics']
data_Type_2_Diabetes_Cases = data['Type 2 Diabetes Cases']
```

```python
# Get mean, median, mode, variance and standard deviation
mean=data_controls.mean()
median=data_controls.median()
mode=data_controls.mode().get_values()[0]
var = data_controls.var()
std = data_controls.std()

# Print out them
print('mean: '+str(mean))
print('median: '+str(median))
print('mode: '+str(mode))
print('variance: '+str(var))
print('standard deviation: '+str(std))

# Create two subplots, one is box plot and one is distribution plot
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.4, 1)},figsize=(12,6))

# Assign different style and color of lines and show them in both plots
sns.boxplot(data_controls,ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

sns.distplot(data_controls,ax=ax_hist,bins=10)
ax_hist.axvline(mean, color='r', linestyle='--')
ax_hist.axvline(median, color='g', linestyle='-')
ax_hist.axvline(mode, color='b', linestyle='-')
ax_hist.legend({'Mode':mode,'Mean':mean,'Median':median})
```
